## Initialize model

Adjust `lookback_months` if needed before running training.

```python
# Auto-load the latest trained .pkl model
model = CyberAttackEvolutionModel.auto_load()
print(f'Loaded model — {len(model.historical_data)} categories, trained={model.trained}')
```

## Train model

Run training and inspect returned metrics.

```python
if getattr(model, 'trained', False):
    print('Model already trained — skipping training. Load from .pkl for faster results.')
    print('To retrain, run: model.train()')
else:
    metrics = model.train()
    print('Training metrics:')
    print(json.dumps(metrics, indent=2))
```

## Save trained model to pickle

Save the trained state for reuse in inference notebooks or production.

```python
out_path = model.save_model()
print('Saved model to:', out_path)
```

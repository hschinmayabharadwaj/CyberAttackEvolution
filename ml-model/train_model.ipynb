```python
import json
import sys
from pathlib import Path

# Ensure the local `ml-model` package is importable when running from the repository root
sys.path.append(str(Path.cwd() / "ml-model"))

from models.evolution_model import CyberAttackEvolutionModel
# Alternatively, install the package with `pip install -e ml-model` if preferred
# Note: If running from repo root, ensure Python path resolves ml-model package or run `pip install -e ml-model`
```

## Initialize model

Adjust `lookback_months` if needed before running training.

```python
model = CyberAttackEvolutionModel(lookback_months=24)
print('Loaded historical categories:', len(model.historical_data))
```

## Train model

Run training and inspect returned metrics.

```python
metrics = model.train()
print('Training metrics:')
print(json.dumps(metrics, indent=2))
```

## Save trained model to pickle

Save the trained state for reuse in inference notebooks or production.

```python
out_path = model.save_model()
print('Saved model to:', out_path)
```

## Load from pickle (example)

You can load the saved model in another notebook or script using `CyberAttackEvolutionModel.load_from_pickle(path)`.

```python
# Example: load the saved model and run a quick prediction
import sys
from pathlib import Path
sys.path.append(str(Path.cwd() / "ml-model"))
from models.evolution_model import CyberAttackEvolutionModel as Loader
loaded = Loader.load_from_pickle(str(out_path))
print('Loaded model categories:', len(loaded.historical_data))
preds = loaded.predict(6)
print('Generated', len(preds), 'predictions')
```
